**Equation of line**
A line is determined by its slope and its intercept. In other words, for each point y on a line we can say: y=mx+x. Where m is the slope, and b is the intercept. y is a given point on the y-axis, and it corresponds to a given x on the x-axis. The slope is a measure of how steep the line is, while the intercept is a measure of where the line hits the y-axis.
